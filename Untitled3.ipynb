```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
# 2. Load Dataset
# Read the PlantGrowth dataset
df = sm.datasets.get_rdataset("PlantGrowth").data

# Display first few rows
print("Dataset Preview:")
print(df.head())
```

```python
# 3. Data Preprocessing
# Check missing values
print("\nMissing Values Check:")
print(df.isnull().sum())

# Summary Statistics
print("\nSummary Statistics:")
print(df.describe())

# Check class balance
print("\nGroup Counts:")
print(df['group'].value_counts())
```

```python
# 4. Assumptions Checking

# 4.1 Normality - Shapiro-Wilk Test
print("\nShapiro-Wilk Test for Normality:")
for grp in df['group'].unique():
    stat, p = stats.shapiro(df[df['group'] == grp]['weight'])
    print(f"Group: {grp}, p-value = {p:.4f}")

# 4.2 Homogeneity of Variance - Levene's Test
stat, p = stats.levene(
    df[df['group'] == 'ctrl']['weight'],
    df[df['group'] == 'trt1']['weight'],
    df[df['group'] == 'trt2']['weight']
)
print(f"\nLevene’s Test for Equal Variance: p-value = {p:.4f}")
```

```python
# 5. Model Building
# Fit the linear model
model = smf.ols('weight ~ group', data=df).fit()
```

```python
# 6. Type I ANOVA
anova_type1 = anova_lm(model, typ=1)
print("\nType I ANOVA Results:")
print(anova_type1)
```

```python
# 7. Type II ANOVA
anova_type2 = anova_lm(model, typ=2)
print("\nType II ANOVA Results:")
print(anova_type2)
```

```python
# 8. Post-Hoc Analysis: Tukey HSD
print("\nTukey HSD Post-Hoc Test Results:")
tukey = pairwise_tukeyhsd(endog=df['weight'], groups=df['group'], alpha=0.05)
print(tukey.summary())

# Plotting Tukey's Confidence Intervals
tukey.plot_simultaneous()
plt.title("Tukey HSD Confidence Intervals")
plt.show()
```

```python
# 9. Effect Size Calculation (Eta Squared)
ss_total = np.sum((df['weight'] - df['weight'].mean())**2)
ss_between = anova_type2['sum_sq'][0]
eta_squared = ss_between / ss_total
print(f"\nEffect Size (Eta Squared): {eta_squared:.4f}")
```

```python
# 10. Visualization
# 10.1 Boxplot
sns.boxplot(x='group', y='weight', data=df, palette='Set2')
plt.title("Boxplot of Plant Weights by Group")
plt.show()

# 10.2 Histogram
sns.histplot(df['weight'], kde=True, color='skyblue')
plt.title("Histogram of Plant Weights")
plt.xlabel("Weight")
plt.ylabel("Frequency")
plt.show()

# 10.3 Violin Plot
sns.violinplot(x='group', y='weight', data=df, palette='muted')
plt.title("Violin Plot of Weights by Group")
plt.show()

# 10.4 Mean Bar Plot
group_means = df.groupby('group')['weight'].mean().reset_index()
sns.barplot(x='group', y='weight', data=group_means, palette='pastel')
plt.title("Mean Plant Weight by Group")
plt.ylabel("Mean Weight (g)")
plt.show()

# 10.5 Q-Q Plot (Normality Check)
sm.qqplot(model.resid, line='45')
plt.title("Q-Q Plot of Residuals")
plt.show()

# 10.6 Residuals vs Fitted
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, line_kws={'color': 'red'})
plt.axhline(0, linestyle='--', color='black')
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted Values")
plt.show()

# 10.7 Density Plot of Residuals
sns.kdeplot(model.resid, fill=True)
plt.title("Density Plot of Residuals")
plt.xlabel("Residuals")
plt.show()

# 10.8 Correlation Heatmap (even if simple for this dataset)
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()
```
